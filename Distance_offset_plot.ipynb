```python
#Here, I am plotting offset vs distance with positive and negative error included

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Plot_CSV.csv")

x = df.iloc[:, 0]        # distance
lat = df.iloc[:, 1]      # mean lateral offset
pos_err = df.iloc[:, 3]  # positive lateral error
neg_err = df.iloc[:, 4]  # negative lateral error

lat_yerr = [neg_err, pos_err]

plt.figure(figsize=(9,5))

plt.errorbar(
    x, lat,
    yerr=lat_yerr,
    fmt="o",
    color="black",      # dot color
    ecolor="0.6",      # error bar color
    capsize=3
)

plt.xlabel("Distance (m)")
plt.ylabel("Lateral offset (m)")
plt.title("Lateral Offset vs Distance for the SCIF")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("/drive/notebooks/figures/lateral_offset_plot.png", dpi=300, bbox_inches="tight")

plt.show()
```
